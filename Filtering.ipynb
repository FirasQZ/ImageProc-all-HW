# Smoothing Images ( Image Filtering )

Image filtering is used to:
1.Remove noise
2.Sharpen contrast
3.Highlight contours
4.Detect edges

low-pass filters(LPF), high-pass filters(HPF) etc. LPF helps in removing noises, blurring the images etc. HPF filters helps in finding edges in the images.

# averaging filter (LPF)...
Average (or mean) filtering is a method of ‘smoothing’ images by reducing the amount of intensity variation between neighbouring
pixels.

How the avg work ?
The average filter works by moving through the image pixel by pixel, replacing each value with the average value of neighbouring pixels, including itself.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('images/5.jpg')
kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(img,-1,kernel)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('images/5.jpg')
blur = cv2.blur(img,(5,5))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

# Median Filtering...
In this type used to remove noise from images and It is widely used as it is very effective at removing noise while
preserving edges.
How the median work ?
The median filter works by moving through the image pixel by pixel, replacing each value with the median value of neighbouring pixels in.(The median is calculated by first sorting all the pixel values from the Matrix into numerical order, and then replacing the pixel in center).

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('images/5.jpg')
median = cv2.medianBlur(img,5)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(median),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```
