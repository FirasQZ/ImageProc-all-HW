```python
import cv2
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt

image = io.imread('Images/mouse.tif',0)
image1 = np.float_(image)
im = ndimage.gaussian_filter(image1, 0)
#sx = ndimage.sobel(im, axis=0)
#sy = ndimage.sobel(im, axis=1)

sx = cv2.Sobel(im,cv2.CV_64F,1,0,ksize=9)
sy = cv2.Sobel(im,cv2.CV_64F,0,1,ksize=9)
M = cv2.multiply(sx,sy)


plt.figure(figsize=(25 , 10))

plt.subplot(151)
plt.imshow(im, 'gray')
plt.axis('off')
plt.title('Orginal', fontsize=20)

plt.subplot(152)
plt.imshow(sx,'gray')
plt.axis('off')
plt.title('Sobel (x direction)', fontsize=20)

plt.subplot(153)
plt.imshow(sy,'gray')
plt.axis('off')
plt.title('Sobel (x direction)', fontsize=20)


plt.subplot(154)
plt.imshow(M,'gray')
plt.axis('off')
plt.title('Sobel Y and X direction)', fontsize=20)

ret,thresh1 = cv2.threshold(M,127,255,cv2.THRESH_BINARY)
plt.subplot(155)
plt.imshow(thresh1,'gray')
plt.axis('off')
plt.title('threshold', fontsize=20)




plt.show()
```
